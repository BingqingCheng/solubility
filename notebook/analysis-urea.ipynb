# $k_BT \sum \ln(\hbar \omega/k_B T)$

```python
def get_Ahar(eos, T):
    return hartree_2_kcalmol*T*kb*np.log(np.sqrt(eos['harmonic_eigval'][:])/(T*kb)).sum() / eos['N']
```

```python
def get_Ahar_Trange(eos, Trange):
    A_har = np.zeros((len(Trange),2))
    for i, T in enumerate(Trange):
        A_har[i,0] = T
        A_har[i,1] = get_Ahar(eos, T)
    eos['ahar'] = A_har
    return A_har
```

```python
def get_zpe(eos):
    return hartree_2_kcalmol*np.sqrt(eos['harmonic_eigval'][:]).sum() / eos['N'] / 2.
```

$Z_{cm} =V \left(\dfrac{m k_B T}{2\pi\hbar^2}\right)^{3/2}$

```python
def get_Acm(eos, T):
    Z = eos['v0'] * (mm*kb*T/(2*math.pi))**(3/2.)
    A_cm = -kb*T*np.log(Z) * hartree_2_kcalmol / eos['N']
    return A_cm
```

```python
def get_Acm_Trange(eos, Trange):
    A_cm = np.zeros((len(Trange),2))
    for i, T in enumerate(Trange):
        A_cm[i,0] = T
        Z = eos['v0'] * (mm*kb*T/(2*math.pi))**(3/2.)
        A_cm[i,1] = -kb*T*np.log(Z) * hartree_2_kcalmol
    eos['acm'] = A_cm
    return A_cm
```

$    Z_{rot} = \sqrt{\pi}\left( 
    \dfrac{2k_B T}{\hbar^2}\right)^{3/2} \sqrt{I_1 I_2 I_3}. $ 

```python
def get_Arot(eos, T):
    I = eos['I']
    try:
        sigma = eos['sigma']
    except:
        sigma = 1
    Z = np.sqrt(math.pi) * (2*kb*T)**(3/2.) * np.sqrt(I[0]*I[1]*I[2]) / sigma
    A_rot = -kb*T*np.log(Z) * hartree_2_kcalmol
    return A_rot
```

```python
def get_Arot_Trange(eos, Trange):
    I = eos['I']
    A_rot = np.zeros((len(Trange),2))
    
    try:
        sigma = eos['sigma']
    except:
        sigma = 1
        
    for i, T in enumerate(Trange):
        A_rot[i,0] = T
        Z = np.sqrt(math.pi) * (2*kb*T)**(3/2.) * np.sqrt(I[0]*I[1]*I[2]) / sigma
        A_rot[i,1] = -kb*T*np.log(Z) * hartree_2_kcalmol
    eos['arot'] = A_rot
    return A_rot
```

# lambda integration

```python
def get_Aanhar_full(eos):
    from scipy.interpolate import interp1d
    from scipy.integrate import quad
    
    dmu_ti = eos['lambda-ti-T0']

    int_func = interp1d(dmu_ti[:,0], dmu_ti[:,1], kind='linear',fill_value='extrapolate')
    int_error_func = interp1d(dmu_ti[:,0], dmu_ti[:,2], kind='linear',fill_value='extrapolate')

    # integrate
    l_mu = []
    n_sample = 0
    l0 = 1
    mu0 = 0
    for i,l in enumerate(dmu_ti[:,0]):
        n_sample+=1
        mu_now = quad(int_func, l0, l)[0] + mu0
        if n_sample >= 2:
            mu_error_now = quad(int_error_func, l0, l)[0]/(n_sample-1)**0.5
        else:
            mu_error_now = quad(int_error_func, l0, l)[0]
        l_mu.append([l,mu_now, mu_error_now])
    return np.asarray(l_mu)
```

```python
def get_Aanhar(eos):
    from scipy.interpolate import interp1d
    from scipy.integrate import quad
    
    dmu_ti = eos['lambda-ti-T0']
    n_sample = len(dmu_ti)

    int_func = interp1d(dmu_ti[:,0], dmu_ti[:,1], kind='linear',fill_value='extrapolate')
    int_error_func = interp1d(dmu_ti[:,0], dmu_ti[:,2], kind='linear',fill_value='extrapolate')

    # integrate
    [l0, l] = [0, 1]
    mu = quad(int_func, l0, l)[0]
    if n_sample >= 2:
        mu_error = quad(int_error_func, l0, l)[0]/(n_sample-1)**0.5
    else:
        mu_error = quad(int_error_func, l0, l)[0]
    return [mu, mu_error]
```

```python
def get_G_T0(eos):
    Ahar = get_Ahar(eos, eos['T0'])
    try:
        Arot = get_Arot(eos, eos['T0'])
    except:
        Arot = 0
    Acm = get_Acm(eos, eos['T0'])
    Aanhar = get_Aanhar(eos)
    eos['g0'] = [Aanhar[0] + Ahar + Arot + Acm, Aanhar[1] ]
    return eos['g0']
```

# Thermodynamic integration along T, in a straight line in Log(T)
### $y=\ln(T/T_0)$
###    $\dfrac{G{(P,T_1)}}{T_1} - \dfrac{G{(P,T_0)}}{T_0} 
    =-\int_0^{\ln(T/T_0)} \left<H\right>_{P,T}/T dy$

```python
def get_dmu_dT(eos, T_range):
    dhnptlist = eos['h']
    
    ttnpt = dhnptlist[:,0]
    dh = dhnptlist[:,1] - eos['u0'] - ttnpt * kb * hartree_2_kcalmol* (3/2.) * eos['n']
    dherror = dhnptlist[:,2]
    H_T = interpolate.interp1d(np.log(ttnpt/eos['T0']), dh/ttnpt)
    H_T_error = interpolate.interp1d(np.log(ttnpt/eos['T0']), dherror/ttnpt)

    Gnow = np.zeros((len(T_range),3))
    for i,T_now in enumerate(T_range):
        Gnow[i,0] = T_now
        Gnow[i,1] = eos['u0'] + T_now/eos['T0'] * (eos['g0'][0]-eos['u0']) \
                   - T_now * kb * hartree_2_kcalmol * 3 * eos['n'] * np.log(T_now/eos['T0']) \
                   - T_now * integrate.quad(H_T, 0, np.log(T_now/eos['T0']))[0]
        # I think the error should be scaled
        Gnow[i,2] = T_now/eos['T0'] * eos['g0'][1] \
                   + T_now * np.abs(integrate.quad(H_T_error, 0, np.log(T_now/eos['T0']))[0])
    eos['g'] = Gnow
    return Gnow
```

```python
# get (ideal) solubility

conv = 1./ (eos_dict['gas']['v0']/(A_2_bohr)**3.) / A3_2_L / avogadro
print(conv)
print("")

delta_mu = {}
for gsol in Gsol_water:
    T_now = gsol[0]
    gex = gsol[1] + G_gas_func(gsol[0]) - G_crystalI_func(gsol[0])
    gex_error = gsol[2]
    frac_est = np.exp(-gex/(kb * hartree_2_kcalmol* T_now))
    #print(T_now, gex, gex_error, "frac:", frac_est, "mol/L", frac_est*conv, "g/L", m_amu*frac_est*conv )
    print("%4.2f g_ex[kcal/mol]: %4.4f %4.4f frac: %4.4f mol/L: %4.4f g/L: %4.4f" %(T_now, gex, gex_error, frac_est, frac_est*conv, m_amu*frac_est*conv))
    delta_mu[T_now] = [gex, gex_error]
```

```python
def ideal_mu(c, c0, kbt):
    return kbt*np.log(c/c0)
```

```python
mu_sol_ex = {}
mu_sol_ex_func = {}
for s in [20, 40, 60, 80, 100]:
    mu_sol_ex[s] = np.loadtxt('./urea-water/md-charmm/mu-ex-T-'+str(s)+'.dat')
    mu_sol_ex[s] = np.vstack([mu_sol_ex[s], np.array([0.000001,0,0])])
    mu_sol_ex_func[s] = interpolate.interp1d(mu_sol_ex[s][:,0],mu_sol_ex[s][:,1])
```
